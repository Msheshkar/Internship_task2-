```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data.csv")
df.head()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(df['Column2'], df['Column4'], marker='o', color='teal')

plt.title("Salary Trend of Employees")
plt.xlabel("Name")
plt.ylabel("Salary")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df['Column2'], df['Column4'], color='plum')
plt.title("Salary Distribution by Name")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.plot(df['Column3'], df['Column4'], marker='o', color='orchid')
plt.title("Salary Trend by Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
df = pd.read_csv("data.csv")

state_counts = df['Column6'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Employee Distribution by State")
plt.show()
```
